# Passo 1: Importar bibliotecas

```python
import pandas as pd
import tensorflow as tf
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
pd.set_option('display.max_columns', None) 
pd.set_option('display.max_rows', None)
```

# Passo 2: Carregar e preparar os dados

```python
df_test = pd.read_csv('iris-test.data', names=['A', 'B', 'C', 'D', 'E'])
df_train = pd.read_csv('iris-train.data', names=['A', 'B', 'C', 'D', 'E'])
```

```python
df_test.info()
```

```python
df_train.info()
```

```python
# Separar atributos e classes
X_train = df_train[['A', 'B', 'C', 'D']]
y_train = df_train['E']
X_test = df_test[['A', 'B', 'C', 'D']]

# Codificar as classes para valores numéricos
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)
```

# Passo 3: Construir o Modelo MLP com TensorFlow

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(10, activation='relu', input_shape=(4,)),
    tf.keras.layers.Dense(10, activation='relu'),
    tf.keras.layers.Dense(3, activation='softmax')
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

model.fit(X_train, y_train_encoded, epochs=200) #epochs=200 são 200 loops do modelo por todo o df
```

# Passo 4: Avaliar o Modelo

```python
# Avaliar o modelo nos dados de teste
predictions = model.predict(X_test)
predicted_classes = label_encoder.inverse_transform(predictions.argmax(axis=-1))
print(predicted_classes)
```

```python
# Extrair rótulos reais dos dados de teste
y_test = label_encoder.transform(df_test['E'])
predicted_labels = predictions.argmax(axis=1)

# Calcular a acurácia
accuracy = accuracy_score(y_test, predicted_labels) #sklearn accuracy_score function
print(f'Acurácia do modelo: {accuracy}')
```
